```python
# Okruženje i biblioteke
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
# Učitavanje ulaza
text = (
"""Elon Musk is the CEO of Tesla and SpaceX. Tesla was founded in 2003 in California.
SpaceX launched the Falcon 9 rocket from Florida. Elon Musk often talks about Mars colonization.
NASA collaborates with SpaceX on several missions. Tesla is also developing autonomous vehicles."""
)
```

```python
# Segmentacija na rečenice
from pprint import pprint

doc = nlp(text)

sentences = [sent.text for sent in doc.sents]

pprint(sentences)
```

```python
# NER po rečenicama
entities = []

for sent in sentences:
    doc_sent = nlp(sent)
    for ent in doc_sent.ents:
        entities.append(ent.text)

pprint(entities)
```

```python
# Statistika: najfrekventniji entiteti
from collections import Counter

entity_freq = Counter(entities)

pprint(entity_freq)

top_entities = [ent for ent, freq in entity_freq.most_common(3)]

print("Najfrekventniji entiteti:", top_entities)
```

```python
# Pronalaženje rečenica koje sadrže te entitete
summary_sentences = []

for sent in sentences:
    for ent in top_entgitities:
        if ent in sent:
            summary_sentences.append(sent)
            break
```

```python
# Generisanje sažetka
summary = " ".join(summary_sentences)

print("SAŽETAK:\n")
print(summary)
```
